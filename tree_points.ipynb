```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import csv
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

nba_team_data = {
 'Team': [],
 'Matchup': [],
 'Game_date': [],
 'W_L': [],
 'Min': [],
 'Pts': [],
 'Fgm': [],
 'Fga': [],
 'Fg_p': [],
 'Three_pm': [],
 'Three_pa': [],
 'Threep_p': [],
 'Ftm': [],
 'Fta': [],
 'Ft_p': [],
 'Oreb': [],
 'Dreb': [],
 'Reb': [],
 'Ast': [],
 'Stl': [],
 'Blk': [],
 'Tov': [],
 'Pf': [],
 'Plus_Minus': []
}

nba_team_prep = {
    'Team1': [],
    'Team2': [],
    ## team1, team2, mean(team1), mean(team2), dif(mean(team1, team 2))
    #
    'Dif_Team1_Team2': [],
    'Win_Loose': [],
    'Point_Diff': []
}

n = pd.read_csv('nba2020.csv')
#format the date and outcome
n['Game_date'] = n['Game_date'].apply(lambda x: datetime.strptime(x.strip(),'%m/%d/%Y'))
n['W_L'] = n['W_L'].apply(lambda x: '1' if x=='W' else '0')
```

```python
#ntp = [[i, j,'','',''] for i in teams for j in teams if i !=j]
teams = n['Team'].unique().tolist()
teams = [x for x in teams]
ntp = nba_team_prep
pairs = [[i, j,'',''] for i in teams for j in teams if i !=j]
```

```python
ntp = [[i,j,n[(n['Team'] == i)].mean(axis = 0) - n[(n['Team'] == j)].mean(axis = 0),'',''] for i in teams for j in teams if i !=j]
```

```python
#Train
from sklearn.tree import DecisionTreeRegressor
#prepare data for win/los
y = n['W_L']
n1 = n.drop('W_L', axis = 1) 
m = n1.to_numpy()
X = m[:,4:]

#prepare data for point difference


#train model for win/loss

tree_reg = DecisionTreeRegressor (max_depth=5)
tree_reg.fit(X, y)

#train model for point difference
mp = n.to_numpy()
X_points = m[:,5:]
y_points = n['Pts']


tree_reg_points = DecisionTreeRegressor (max_depth=15)
lasso_reg_points.fit(X_points, y_points)

```

```python
for game in range(len(pairs)):
    x = (np.asarray(ntp[game][2])[1:]).reshape(1,-1)
    xp = (np.asarray(ntp[game][2])[2:]).reshape(1,-1)
    pairs[game][2] = np.round(lasso_reg.predict(x)[0], decimals=1)
    pairs[game][3] = np.round(lasso_reg_points.predict(xp)[0], decimals=1)
    
```
