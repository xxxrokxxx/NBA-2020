```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import re #need to do some regular expression work
import random
import datetime

#scrap with the assistance from FireFox driver: geckodriver. Make sure geckodriver is in your $PATH. /usr/local/bin
driver = webdriver.Firefox()
driver.implicitly_wait(30)
driver.get("https://stats.nba.com/teams/boxscores/?Season=2019-20&SeasonType=Regular%20Season")

```

```python
soup_level1=BeautifulSoup(driver.page_source, 'lxml') #make sure we can read the data too :) 
```

```python
nba_team_data = {
 'Team': [],
 'Matchup': [],
 'Game_date': [],
 'W_L': [],
 'Min': [],
 'Pts': [],
 'Fgm': [],
 'Fga': [],
 'Fg_p': [],
 'Three_pm': [],
 'Three_pa': [],
 'Threep_p': [],
 'Ftm': [],
 'Fta': [],
 'Ft_p': [],
 'Oreb': [],
 'Dreb': [],
 'Reb': [],
 'Ast': [],
 'Stl': [],
 'Blk': [],
 'Tov': [],
 'Pf': [],
 'Plus_Minus': []
}
```

```python
#move to the next page
#python_button = driver.find_element_by_class_name('stats-table-pagination__next') #desno
#python_button.click() #click right link
#print(soup_level1.prettify())
```

```python
#1st page
games = soup_level1.find_all('td')
```

games

```python
def scrape_page():
    #get through the table and insert into DataFrame
    i = 0
    
    for game in games:
        if len(games[i+1].text) >= 11:
            nba_team_data['Team'].append(games[i].text.strip())
            nba_team_data['Matchup'].append(games[i+1].text.strip())
            nba_team_data['Game_date'].append(games[i+2].text.strip())
            nba_team_data['W_L'].append(games[i+4].text.strip())
            nba_team_data['Min'].append(games[i+5].text)
            nba_team_data['Pts'].append(games[i+6].text)
            nba_team_data['Fgm'].append(games[i+7].text.strip())
            nba_team_data['Fga'].append(games[i+8].text.strip())
            nba_team_data['Fg_p'].append(games[i+9].text)
            nba_team_data['Three_pm'].append(games[i+10].text.strip())
            nba_team_data['Three_pa'].append(games[i+11].text.strip())
            nba_team_data['Threep_p'].append(games[i+12].text)
            nba_team_data['Ftm'].append(games[i+13].text)
            nba_team_data['Fta'].append(games[i+14].text)
            nba_team_data['Ft_p'].append(games[i+15].text)
            nba_team_data['Oreb'].append(games[i+16].text.strip())
            nba_team_data['Dreb'].append(games[i+17].text.strip())
            nba_team_data['Reb'].append(games[i+18].text.strip())
            nba_team_data['Ast'].append(games[i+19].text.strip())
            nba_team_data['Stl'].append(games[i+20].text.strip())
            nba_team_data['Blk'].append(games[i+21].text.strip())
            nba_team_data['Tov'].append(games[i+22].text.strip())
            nba_team_data['Pf'].append(games[i+23].text)
            nba_team_data['Plus_Minus'].append(games[i+24].text)
            i = i + 25
            #print(i)
            #print(len(games[i+1].text))
    
```

```python
#scrape all the 39 pages
j = 0
while j < 40:
    scrape_page()
    #go to next page
    r = random.randint(10,1000)
    driver.implicitly_wait(r)
    python_button = driver.find_element_by_class_name('stats-table-pagination__next') #right arrow
    python_button.click() #click right link
    soup_level2=BeautifulSoup(driver.page_source, 'lxml')
    games = soup_level2.find_all('td')
    j = j+1

```

```python
nba_table = pd.DataFrame(nba_team_data)
```
